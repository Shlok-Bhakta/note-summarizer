```python
import torch  
import whisper 
import os
import subprocess
import gc
import openai
fileStorage = "GeologyW7"
aiSummary = True
openai.api_key = os.getenv("OPENAIKEY")
```

```python
files = os.listdir(fileStorage)
if not os.path.exists(f"{fileStorage}/audio"):
    os.makedirs(f"{fileStorage}/audio")
for i in files:
    print(i)
    inputFile = f"{fileStorage}/{i}"
    outputFile = f"{fileStorage}/audio/{i[:-4]}.mp3"
    subprocess.call(["ffmpeg", "-i", inputFile, "-map", "0:a", outputFile])
```

```python
files = os.listdir(f"{fileStorage}/audio")
if not os.path.exists(f"{fileStorage}/text"):
    os.makedirs(f"{fileStorage}/text")
```

```python
devices = torch.device("cuda:0" if torch.cuda.is_available() else "cpu")
masterFile = open(f"{fileStorage}/text/combined.txt", "w") 
model = whisper.load_model("small" , device =devices)
prompt = "I will give you text. I need a thorough summary. Its fine if the summary is long, just remove some of the fluff and keep all important stuff that might appear on an exam, dont refrence the exam in your output"
if(aiSummary == True):
    aiFile = open(f"{fileStorage}/text/summarized.txt", "w") 

for i in files:
    audiofile = f"{fileStorage}/audio/{i}"
    result = model.transcribe(audiofile)
    print(result)

    f = open(f"{fileStorage}/text/{i[:-4]}.txt", "w")
    f.write(result["text"]+f"\n\n#️⃣#️⃣#️⃣#️⃣#️⃣#️⃣#️⃣#️⃣#️⃣#️⃣#️⃣#️⃣#️⃣#️⃣#️⃣#️⃣#️⃣#️⃣ {i}\n\n")
    masterFile.write(f"\n\n#️⃣#️⃣#️⃣#️⃣#️⃣#️⃣#️⃣#️⃣#️⃣#️⃣#️⃣#️⃣#️⃣#️⃣#️⃣#️⃣#️⃣#️⃣ {i}\n\n" + result["text"])
    f.close()
    
    if(aiSummary == True):
        print("CHATGPT Summarizing...\n")
        result = openai.ChatCompletion.create(
        model="gpt-3.5-turbo",
        messages=[
            {"role": "system", "content": prompt},
            {"role": "user", "content": result["text"]}
        ]
        )
        print("summarized: " + result['choices'][0]['message']['content'] + "\n")
        aiFile.write(f"\n\n# #️⃣#️⃣#️⃣#️⃣#️⃣#️⃣#️⃣#️⃣#️⃣#️⃣#️⃣#️⃣#️⃣#️⃣#️⃣#️⃣#️⃣#️⃣ {i}\n\n" + result['choices'][0]['message']['content'])
        
masterFile.close()
aiFile.close()
```

```python
del model
gc.collect()
torch.cuda.empty_cache()
```
